```python
import os
from dotenv import load_dotenv
load_dotenv()

os.environ["GOOGLE_API_KEY"] = os.getenv("GOOGLE_API_KEY")
```

```python
from langchain.chat_models import init_chat_model

model = init_chat_model("google_genai:gemini-2.5-flash")

res = model.stream("help me to get the weather of hyderabad")
res
```

```python
for chunk in model.stream("what is artificial intellengence AI"):
    print(chunk.content,flush = True)
    
```

```python
##batch processing 

res= model.batch([
    "what is ai",
    "what is fastApi",
    "what is langchain"
])

for response in res:
    print(response.content)
```
