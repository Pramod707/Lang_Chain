```python
import os
from dotenv import load_dotenv
from langchain.chat_models import init_chat_model
load_dotenv()

os.environ["GOOGLE_API_KEY"] = os.getenv("GOOGLE_API_KEY")

model = init_chat_model("google_genai:gemini-2.5-flash")

```

```python
from langchain.tools import tool

@tool
def get_weather(city:str)->str:
    """get the weather of the city"""
    return f"the weather of {city}is sunny"

tools = model.bind_tools([get_weather])

res = tools.invoke("whats the weather in hyderabad")
print(res.content)
print(res.tool_calls)

# for tool in tools:
#     print(f"tool : {tool[0]}")
#     print(f"args: {tool[1]}")
    
    
```

```python
message = [{
    "role" : "user",
    "content" : "what is the weather of hyderabad"
}]
ai_msg = tools.invoke(message)
message.append(ai_msg)

for tool_call in ai_msg.tool_calls:
    tool_res = get_weather.invoke(tool_call)
    message.append(tool_res)

final_res = tools.invoke(message)
print(final_res.content)
    


```

```python
from typing_extensions import TypedDict , Annotated

class Movie(TypedDict):
    name : Annotated[str,...,"Name of the movie"]
    year : Annotated[str,...,"Year of the movie"]
    genre : Annotated[str,...,"Genre of the movie"]
    rating : Annotated[float,...,"Rating of the movie"]
    collection : Annotated[int,...,"Collection of the movie"]
    
    
res = model.with_structured_output(Movie)

res.invoke("provide me details of movie bahubali")
res
```
