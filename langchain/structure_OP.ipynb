```python
import os
from dotenv import load_dotenv
from langchain.chat_models import init_chat_model
load_dotenv()

os.environ["GOOGLE_API_KEY"] = os.getenv("GOOGLE_API_KEY")

model = init_chat_model("google_genai:gemini-2.5-flash")
model
```

```python
from pydantic import BaseModel, Field

class Movie(BaseModel):
    name:str=Field(description= "Name of the movie")
    year:int=Field(description= "Year of the movie")
    genre:str=Field(description= "Genre of the movie")
    rating : float=Field(description="Rating of the movie")
    director:str=Field(description= "Director of the movie")
```

```python
model_with_structuredOP = model.with_structured_output(Movie)
model_with_structuredOP
```

```python
model.invoke("Provide me details of movie bahubali")
```

```python
res = model_with_structuredOP.invoke("provide me details of movie bahubali")
print(res)
```
